```python
def regex_func(txt):
    txt = re.findall('(?<!\S)[A-Za-z]+(?!\S)|(?<!\S)[A-Za-z]+(?=:(?!\S))',txt)
    txt = [word for word in txt if word.lower() not in stop_words]
    txt = re.findall('[A-Z]+[^A-Z]*', ' '.join(txt))
    txt = [word.lower() for word in (' '.join(txt)).split()]
    return txt
```

```python
for col in trained_words.columns:
    trained_words.loc[:,col] = trained_words[col].apply(regex_func)
```

```python
def combining_columns(df):
    columns = df.columns
    df['combined'] = np.empty((len(df), 0)).tolist()
    for col in columns:
        df['combined'] = df['combined'] + df[col]
    
    return df
```

```python
trained_words = combining_columns(trained_words)
```

```python
lemmatizer = WordNetLemmatizer()
trained_words.loc[:,'combined'] = trained_words['combined'].map(lambda x : [lemmatizer.lemmatize(y) for y in x])
```

```python
trained_words['cleaned_text'] = trained_words['combined'].apply(lambda x : ' '.join(x))
```

```python
cvec =CountVectorizer(max_features=50000,ngram_range=(1,1))
```

```python
word_vec = pd.concat([df['fraudulent'],trained_words['cleaned_text']], axis = 1)
```

```python
cvec.fit(word_vec.loc[:,'cleaned_text'])
word_counts = cvec.fit_transform(word_vec['cleaned_text'])
```
